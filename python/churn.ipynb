```python
features = df.drop('Exited', axis=1)
```

```python
features
```

```python
target = df["Exited"]
```

```python
target
```

```python
features = features.drop(["RowNumber", "CustomerId", "Surname"], axis=1)
```

```python
features = features.dropna()
features
```

```python
features = pd.get_dummies(features, columns=["Geography", "Gender"])
```

```python
features
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
scalar = StandardScaler()

x_train = scalar.fit_transform(x_train)
x_test = scalar.fit_transform(x_test)
```

```python
x_train
x_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pickle5 as pickle

import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
lr_model = LogisticRegression(random_state=42)
```

```python
lr_model.fit(x_train, y_train)
```

```python
lr_perdictions = lr_model.predict(x_test)
```

```python
lr_perdictions
```

```python
lr_accuracy = accuracy_score(y_test, lr_perdictions)
```

```python
lr_accuracy
```

```python
def evaluate_and_save_model(model, x_train, x_test, y_train, y_test, filename):
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuraccy = accuracy_score(y_test, y_pred)
    print(f"{model.__class__.__name__} Accurracy: {accuraccy:.4f}")
    print(f"\nClassifacation Report\n{classification_report(y_test, y_pred)}\n\n ------------------------------------ \n")

    with open(filename, "wb") as file:
        pickle.dump(model, file)

    print(f"\nModel Saved as {filename}")

```

```python
xgb_model = xgb.XGBClassifier(random_state=42)
evaluate_and_save_model(xgb_model, x_train, x_test, y_train, y_test, "xgb_modle.pkl")
```

```python
dt_model = DecisionTreeClassifier(random_state = 42)
evaluate_and_save_model(dt_model, x_train, x_test, y_train, y_test, "dt_model.pkl")
```

```python
rf_model = RandomForestClassifier(random_state = 42)
evaluate_and_save_model(rf_model, x_train, x_test, y_train, y_test, "rf_model.pkl")
```

```python
nb_model = GaussianNB()
evaluate_and_save_model(nb_model, x_train, x_test, y_train, y_test, "nb_model.pkl")
```

```python
knn_model = KNeighborsClassifier()
evaluate_and_save_model(knn_model, x_train, x_test, y_train, y_test, "knn_model.pkl")
```
